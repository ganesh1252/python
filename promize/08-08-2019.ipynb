```python
import pandas as pd
df=pd.read_csv("Cars1.csv")
df
```

```python
Y=df.iloc[:,1:2]
X=df.iloc[:,[0,2,3,]]
X
```

```python
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df.iloc[:,:])
```

```python
import statsmodels.formula.api as smf
ml=smf.ols("MPG ~ HP + VOL + SP + WT",data=df).fit()
ml.params
```

```python
from sklearn.metrics import r2_score as r
score=r(X,Y)
```

```python
ml.summary()
```
